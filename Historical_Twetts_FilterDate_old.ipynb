```python
##PARTE 1 - BIBLIOTECAS
import os, sys, re, getopt
import traceback
import pandas as pd
import GetOldTweets3 as got
#pip install GetOldTweets3
```

```python
##PARTE 2 - CRITERIOS DE BUSCA
#set search variables
keyword = "#PETR4"
oldest_date = "2019-01-06"    
newest_date = "2019-01-09"
number_tweets = 0     #0 pega tudo 

tweet_criteria = got.manager.TweetCriteria().setQuerySearch(keyword)\
                                            .setSince(oldest_date)\
                                            .setUntil(newest_date)\
                                            .setMaxTweets(number_tweets)

tweet_df = pd.DataFrame({'got_criteria':got.manager.TweetManager.getTweets(tweet_criteria)})

#tweet_df.shape
#tweet_df.to_csv("c:/app/teste.csv")
#tweet_df
```

```python
##PARTE 3 - CRITERIOS DE BUSCA 
# OBS: CASO NAO TRAGA NADA, RODE ANTES A PARTE 4  E DPS VOLTE NESSA PARTE 3
#extract twitter information
get_twitter_info()
tweet_df = tweet_df.drop("got_criteria", 1)
tweet_df.head()
```

```python
##PARTE 4 - DEFINICAO DE CRITERIOS
def get_twitter_info():
    tweet_df["Text"] = tweet_df["got_criteria"].apply(lambda x: x.text)
    tweet_df["Data"] = tweet_df["got_criteria"].apply(lambda x: x.date)
    tweet_df["Hashtags"] = tweet_df["got_criteria"].apply(lambda x: x.hashtags)
    tweet_df["Link"] = tweet_df["got_criteria"].apply(lambda x: x.permalink)
tweet_df.shape
```

```python
##PARTE 5 - CRIAR UM NOVO DF COM NOVOS DADOS
df_twitter = tweet_df
tweet_df.shape

# caso precise concatenar novo df com ja existente
#df_twitter = pd.concat([df_twitter, tweet_df]) 
```
